# Módulo 2 - Aula 5

```
##substituição por média
avg_1 = df['normalized-losses'].astype('float').mean(axis = 0) 
df['normalized-losses'].replace(np.nan, avg_1, inplace = True)

avg_2=df['bore'].astype('float').mean(axis=0)
df['bore'].replace(np.nan, avg_2, inplace= True)

avg_3 = df["stroke"].astype("float").mean(axis = 0)
df["stroke"].replace(np.nan, avg_3, inplace = True)

avg_4=df['horsepower'].astype('float').mean(axis=0)
df['horsepower'].replace(np.nan, avg_4, inplace= True)

avg_5=df['peak-rpm'].astype('float').mean(axis=0)
df['peak-rpm'].replace(np.nan, avg_5, inplace= True)

##valores que não são números, substituição por frequência'
df['num-of-doors'].value_counts()
df["num-of-doors"].replace(np.nan, "four", inplace = True)
```

```
###eliminar coluna, caso não seja possível nenhuma das premissas.###
df.dropna(subset=["price"], axis=0, inplace = True)

# reset index, because we droped two rows
df.reset_index(drop = True, inplace = True)
```

```
df.isnull().sum()
```

```
%matplotlib inline
df.replace("?",np.nan,inplace=True)
plt.figure(figsize=[30,10])
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap="Blues")
sns.set(font_scale=1)
plt.show()
#plt.close()
```

```
##ajustar tipos de colunas
df.dtypes
```

```
df.head()
```

```
df[['bore', 'stroke']] = df[['bore', 'stroke']].astype('float')
df[['normalized-losses']] = df[['normalized-losses']].astype('int')
df[['price']] = df[['price']].astype('float')
df[['peak-rpm']] = df[['peak-rpm']].astype('float')
df.dtypes
```

# Módulo 3 - Aula 7

```
# Estandardização: transformar mpg a L/100km  (235 dividido por  )
df['city-L/100km'] = 235/df['city-mpg']
df['highway-L/100km'] = 235/df['highway-mpg']

df[['city-mpg','city-L/100km','highway-mpg','highway-L/100km']].head()
```

```
#normalização - converter na escala de 0 a 1
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max()
df[["length","width","height"]].head()
```

```
##Discretização - vamos pegar 4 escalas, logo 3 categorias
df["horsepower"]=df["horsepower"].astype(float, copy=True)
binwidth = (max(df["horsepower"])-min(df["horsepower"]))/4
bins = np.arange(min(df["horsepower"]), max(df["horsepower"]), binwidth)
bins
```

```
group_names = ['Low', 'Medium', 'High']
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names,include_lowest=True )
df[['horsepower','horsepower-binned']].head(20)
#######vamos ver o resultado gráfico
%matplotlib inline
from matplotlib import pyplot as plt
a = (0,1,2)
# draw historgram of attribute "horsepower" with bins = 3
plt.hist(df["horsepower"], bins = 3)
# set x/y labels and plot title
plt.xlabel("Horsepower")
plt.ylabel("Count")
plt.title("Horsepower Bins")
plt.show()
```



# Módulo 3 - Aula 8

```
##conversão de variáveis qualitativas em quantitativas
dummy_fuel = pd.get_dummies(df["fuel-type"])
dummy_fuel.rename(columns={'gas':'fuel-type-gas', 'diesel':'fuel-type-diesel'}, inplace=True)
dummy_fuel.head()
##adiciona essas 2 colunas em nosso dataset e remove coluna anterior

# merge data frame "df" and "dummy_fuel" 
df = pd.concat([df, dummy_fuel], axis=1) #axis=1 quer dizer coluna

# drop original column "fuel-type" from "df"
df.drop("fuel-type", axis = 1, inplace=True) #axis=1 quer dizer coluna
df.head()

##vamos repetir para a coluna aspiration
# get indicator variables of aspiration and assign it to data frame "dummy_variable_2"
dummy_aspiration = pd.get_dummies(df['aspiration'])
# show first 5 instances of data frame "dummy_variable_1"
dummy_aspiration.head()

# change column names for clarity
dummy_aspiration.rename(columns={'std':'aspiration-std', 'turbo': 'aspiration-turbo'}, inplace=True)
# merge data frame "df" and "dummy_variable_1" 
df = pd.concat([df, dummy_aspiration], axis=1)

# drop original column "fuel-type" from "df"
df.drop('aspiration', axis = 1, inplace=True)
df.head()
```

# Módulo 4 - Aula 12

```
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X = df[['engine-size']]
Y = df['price'].values

X_train,X_test,Y_train,Y_test=train_test_split(X,Y, test_size=0.2,random_state=0) # proporção da divisão, o random_state garante o mesmo tipo de sorteio

##verificando separação
X_train.describe()
X_test.describe()
```

```
regressor = LinearRegression()
regressor.fit(X_train,Y_train)
```

```
#visualização dados de training
regressor_train=plt
regressor_train.scatter(X_train, Y_train, color='blue')
y_predict=regressor.predict(X_train)
regressor_train.plot(X_train,y_predict,color='black')
regressor_train.title('Preço por tamanho do motor (Training)')
regressor_train.xlabel('Tamanho do Motor')
regressor_train.ylabel('Preço')
regressor_train.show()
```

```
#visualização dados de teste
regressor_train=plt
regressor_train.scatter(X_test, Y_test, color='blue')
y_predict = regressor.predict(X_train)
regressor_train.plot(X_train,y_predict,color='black')
regressor_train.title('Preço por tamanho do motor (Test)')
regressor_train.xlabel('Tamanho do Motor')
regressor_train.ylabel('Preço')
regressor_train.show()
```

```
#função definida
print("Intercepto: {},Coeficiente: {}".format(regressor.intercept_,regressor.coef_))
```

```
regressor.score(X_test,Y_test)
```

# Módulo 4 - Aula 13

```
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X = df[['horsepower','curb-weight','engine-size','length','width']]
Y = df['price'].values

X_train,X_test,Y_train,Y_test=train_test_split(X,Y, test_size=0.2,random_state=0)

##verificando separação
X_train.describe()
X_test.describe()
```

```
regressor = LinearRegression()
regressor.fit(X_train,Y_train)
```

```
#verificação de gráfico
fig, ax =plt.subplots(figsize=[5,5])
ax = sns.distplot(Y_train,hist=False,color='b',ax=ax)
y_predict=regressor.predict(X_train)
sns.distplot(y_predict,color='r',ax=ax)

plt.legend(labels=['Price','Price - Predict (Training)'])
plt.show()
```

```
regressor.score(X_test,Y_test)
```

```
#função definida
print("Intercepto: {},Coeficiente: {}".format(regressor.intercept_,regressor.coef_))
```

# Módulo 4 - Aula 14

```
from sklearn.ensemble import RandomForestRegressor

X = df[['horsepower','curb-weight','engine-size','length','width']]
Y = df['price'].values

X_train,X_test,Y_train,Y_test=train_test_split(X,Y, test_size=0.2,random_state=0)

##verificando separação
X_train.describe()
X_test.describe()
```

```
forest = RandomForestRegressor()
forest.fit(X_train,Y_train)
```

```
#verificação de gráfico
fig, ax =plt.subplots(figsize=[5,5])
ax = sns.distplot(Y_train,hist=False,color='b',ax=ax)
y_predict=forest.predict(X_train)
sns.distplot(y_predict,color='r',ax=ax)

plt.legend(labels=['Price','Price - Predict (Training)'])
plt.show()
```

```
forest.score(X_test,Y_test)
```
